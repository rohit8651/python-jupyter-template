### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

# Function to create a pie chart
def plot_pie_chart(labels, sizes, colors):
    try:
        # Validate input data
        if not labels or not sizes:
            raise ValueError("Labels and sizes cannot be empty.")
        if len(labels) != len(sizes):
            raise ValueError("Labels and sizes lists must have the same length.")
        if any(size <= 0 for size in sizes):
            raise ValueError("All sizes must be greater than zero.")

        # Create the pie chart
        plt.figure(figsize=(6, 6))  # Set figure size
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
        plt.title('Programming Language Popularity')

        # Display the plot
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")

# Define data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'orange']

# Call the function
plot_pie_chart(labels, sizes, colors)
```
