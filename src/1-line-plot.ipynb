### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to generate plot with error handling
def plot_square_function():
    try:
        # Generate x values and compute y values
        x = np.linspace(0, 10, 100)

        if x is None or len(x) == 0:
            raise ValueError("Input x values are missing or invalid.")

        y = x**2

        # Create the plot
        plt.plot(x, y, label='y = x^2', color='blue')
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.title('Line Plot of y = x^2')
        plt.grid(True)
        plt.legend()

        # Display the plot
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
plot_square_function()
```
