### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot
x = np.linspace(0, 10, 100)
y = x**2
axs[0, 0].plot(x, y, label='y = x^2', color='blue')
axs[0, 0].set_title('Line Plot')
axs[0, 0].grid(True)

# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Plot')

# Scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x, y, color='blue')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'])
axs[1, 1].set_title('Pie Chart')

plt.show()
```
